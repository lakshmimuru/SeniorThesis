```python
#create more Ataxia cal experiments
data = np.load('../datasets/ataxia_cal/data.npy',allow_pickle=True)
mask = np.load('../datasets/ataxia_cal/mask.npy',allow_pickle=True)

all_data = np.load('../datasets/ataxia_reduc/data.npy',allow_pickle=True)
all_mask = np.load('../datasets/ataxia_reduc/mask.npy',allow_pickle=True)

patients = [1,10,20,33,40]

for i, patient in enumerate(patients):
        
    directory = f'../datasets/ataxia_cal_{i}/'
    if not os.path.exists(directory):
        os.makedirs(directory)
    target = all_data[patient]
    mask_target = all_mask[patient]
    present = np.sum((1-mask_target[:,0])>0)
    print(f'non missing values for patient {i}: {present}')
    data_this = data.copy()
    mask_this = mask.copy()
    data_this[0,4:22] = target
    mask_this[0,4:22] = mask_target
    np.save(directory+'data.npy',data_this)
    np.save(directory+'mask.npy',mask_this)

     


    
```
